```python
import os

import tensorflow as tf
from tensorflow import keras

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.image import ImageDataGenerator

from hour_glass_model import glassnet
from utility import show_plot, get_f1_score

physical_devices = tf.config.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(physical_devices[0], True)
print(tf.config.list_physical_devices())
```

```python
# directory
train_dir='./OCT2017_/train/'
test_dir='./OCT2017_/test/'
val_dir='./OCT2017_/val/'
```

```python
ep = 50
early_terminate = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', min_delta=0, patience=int(ep//5), verbose=0,
    mode='auto', baseline=None, restore_best_weights=True
)
```

```python
pic_size=(224, 224)
img_gen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255.0)
tra_generator=img_gen.flow_from_directory(train_dir,target_size=pic_size,batch_size=10,\
                                               class_mode='categorical', interpolation='bilinear')
val_generator=img_gen.flow_from_directory(val_dir,target_size=pic_size,batch_size=10,\
                                               class_mode='categorical', interpolation='bilinear')
tst_generator=img_gen.flow_from_directory(test_dir,target_size=pic_size,batch_size=10,\
                                               class_mode='categorical', interpolation='bilinear')
glass = glassnet(n_classes=4)
print(glass.summary())
glass_history1 = glass.fit_generator(tra_generator,validation_data = val_generator, epochs=ep, verbose=1, callbacks=[early_terminate])
score = glass.evaluate(x=tst_generator)
f1 = get_f1_score(score[2], score[3])
print('f1 {}, score {}'.format(f1, score))
```
